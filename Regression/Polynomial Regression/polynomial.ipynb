```python
import pandas  as pd 
import numpy   as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Age (years) and Height (cm) dataset
age = np.array([5, 7, 10, 12, 15, 17, 20, 22, 25, 28]).reshape(-1, 1)
height = np.array([105, 115, 135, 145, 160, 168, 172, 175, 178, 180])
```

```python
## Step 2: Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(age, height)
height_pred_linear = lin_reg.predict(age)
```

```python
# Step 3: Polynomial Regression
poly = PolynomialFeatures()
poly_age = poly.fit_transform(age)
poly_age
```

```python
model = LinearRegression()
model.fit(poly_age,height)
height_pred_poly = model.predict(poly_age)
```

```python
# Step 4: Plot results
plt.scatter(age,height,color='black',label="actual");plt.title("Actual Age")
plt.plot(age,height_pred_linear,color='b',label="Linear");plt.title("Linear Age")
plt.plot(age,height_pred_poly,color='r',label="Polynomial");plt.title("Linear Age")
plt.legend()
```
